11. Remove outliers from dataset.

```python
q25 = df.quantile(0.25)
q75 = df.quantile(0.75)
iqr = q75 - q25
iqr
```

```python
#show not outliers
df[((df > q25 - 1.5 * iqr) & (df < q75 + 1.5 * iqr)).all(axis=1)]
```

```python
#show outliers 
df[((df < q25 - 1.5 * iqr) | (df > q75 + 1.5 * iqr)).any(axis=1)]
```
